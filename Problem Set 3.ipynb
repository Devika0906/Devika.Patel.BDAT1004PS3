Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
#Step 1
import pandas as pa 
```

```python
#Step 2 & 3 & 4
users = pa.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
print('Mean age per occupation is:')
users.groupby('occupation').age.mean()
```

```python
#Step 5
print('Male ratio per occupation in the most to least form:')
Occupation = users.groupby(['occupation','gender']).gender.count()
Total = users.groupby('gender').gender.count()
print((Occupation/Total).sort_values(ascending=False).xs('M',level=1)) 
```

```python
#Step 6
print('Minimum and Maximum of ages:')
users.groupby('occupation').agg({'ages':['min','max']})   
```

```python
#Step 7
print('Calculated mean age and sex based on occupation:')
users.groupby(['occupation','gender']).age.mean() 
```

```python
#Step 8
Occupations = users.groupby('occupation').gender.count()
gendercount_occupation = users.groupby(['occupation','gender']).gender.count()
print(gendercount_occupation*100/Occupations) 
```
